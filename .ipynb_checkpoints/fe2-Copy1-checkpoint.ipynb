# Exploratory Data Analysis

```python
import numpy as np
import pandas as pd
from sklearn.svm import SVC
from sklearn.cross_validation import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid", palette="muted")
plt.figure(figsize=(10,10))

```

```python
data = pd.read_csv('data.csv')
print "Sample data set"
data.head()
```

```python
print "Features in dataset are: "
print "-------------------------"
for col in data.columns:
    print col
```

### We don't need id and Unnamed:32 column. Drop it!

```python
data = data.drop(['Unnamed: 32','id'], axis=1)
```

### Separate features and target class

```python
y = data.diagnosis #Class M or B
x = data.drop(['diagnosis'], axis=1)
def normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        max_value = df[feature_name].max()
        min_value = df[feature_name].min()
        result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result

x = normalize(x)
x.head()
```

### Number of samples that are M or B

```python
ax = sns.countplot(y,label="Count")       # M = 212, B = 357
B, M = y.value_counts()
print 'Number of Benign: ',B
print'Number of Malignant: ',M
```

```python
drop_list1 = ['perimeter_mean','radius_mean','compactness_mean','concave points_mean','radius_se','perimeter_se','radius_worst','perimeter_worst','compactness_worst','concave points_worst','compactness_se','concave points_se','texture_worst','area_worst']
x = x.drop(drop_list1,axis = 1 )        
x.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
svc = SVC(kernel = 'linear',C=.1, gamma=10, probability = True)
svc.fit(x,y)
y_pred = svc.fit(X_train, y_train).predict(X_test)
t = pd.DataFrame(svc.predict_proba(X_test))
```

```python
i=0
count=0
for index, row in t.iterrows():
    
    if row[0]>row[1] and y_test[i]=='B':
        count+=1
    elif row[0]<row[1] and y_test[i]=='M':
        count+=1
    
    i+=1
print "Accuracy is ", 100.0*count/i
```
